```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

# Functional Method

```python
plt.plot(x,y)

plt.xlabel('X Label')    # add x label
plt.ylabel('Y Label')    # add y label
plt.title('Title')       # add title
```

## create multiple plot on one canvas (subplot)

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])       # [left, bottom, weidth, height]

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x, 'r')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# axes.plot(x,y)
# plt.tight_layout()     # to avoid overlaping

axes[0].plot(x,y,'r')
axes[0].set_title('First')

axes[1].plot(y,x)
axes[1].set_title('Second')

plt.tight_layout()
```

# Figure Size, Aspect Ratio and DPI

```python
fig = plt.figure(figsize=(5,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,4))

axes[0].plot(x,y)

axes[1].plot(y,x)
plt.tight_layout()
```

## SAVE Figure

```python
fig.savefig('my_graph.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Square')
ax.plot(x,x**3, label='X Cube')


ax.legend(loc=0)
```

# Plot Appearance

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,
          color='purple',    # set color of line we can use hex code or common color name eg 'red','blue','green'
          linewidth=2,       # control line width
          alpha=0.6,          # line visibility
          linestyle='-',
          marker='o',        # mark datapoints on line 
          markersize=10,     # set marker size
          markerfacecolor='yellow',
          markeredgewidth=3,
          markeredgecolor='black'
         )
```

## line limit

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,
          color='purple',    # set color of line we can use hex code or common color name eg 'red','blue','green'
          linewidth=2)

axes.set_xlim([0,1])
axes.set_ylim([0,1])
```
